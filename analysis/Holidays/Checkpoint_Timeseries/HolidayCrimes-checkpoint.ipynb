# Comparing Crimes on Avg vs. Holidays

We take the average number of daily crimes per months for the months that have federal holidays in them. Specifically January, February, May, July, September, October, November, and December.


First we read in our reports.out file that includes how many crimes were reported per day:

```python
import sys
import pandas as pd
from datetime import datetime

x=open("reports.out")
crimehistory = list(x)

crimehistory=[i.strip() for i in crimehistory]
crimehistory=[i.split('\t', 1) for i in crimehistory]
```

We find the average daily crimes that happened in each of the months that have federal holidays in.

```python
jan=feb=may=july=sept=octo=nov=dec=0

daily_averages=[]
#Finding Jan Daily Average

for i in crimehistory:
    if i[0][0:2]=='01':
        jan=int(i[1])+jan
jan_daily=int(jan/31/10)
daily_averages.append("January,%s" %(jan_daily))

#Finding Feb Daily Average
for i in crimehistory:
    if i[0][0:2]=='02':
        feb=int(i[1])+feb
feb_daily=int(feb/28/10)
daily_averages.append("February,%s" %(feb_daily))

#Finding May Daily Average
for i in crimehistory:
    if i[0][0:2]=='05':
        may=int(i[1])+may
may_daily=int(may/31/10)
daily_averages.append("May,%s" %(may_daily))

#Finding July Daily Average
for i in crimehistory:
    if i[0][0:2]=='07':
        july=int(i[1])+july
july_daily=int(july/31/10)
daily_averages.append("July,%s" %(july_daily))

#Finding September Daily Average
for i in crimehistory:
    if i[0][0:2]=='09':
        sept=int(i[1])+sept
sept_daily=int(sept/30/10)
daily_averages.append("September,%s" %(sept_daily))

#Finding October Daily Average
for i in crimehistory:
    if i[0][0:2]=='10':
        octo=int(i[1])+octo
octo_daily=int(octo/31/10)
daily_averages.append("October,%s" %(octo_daily))

#Finding November Daily Average
for i in crimehistory:
    if i[0][0:2]=='11':
        nov=int(i[1])+nov
nov_daily=int(nov/30/10)
daily_averages.append("November,%s" %(nov_daily))


#Finding December Daily Average
for i in crimehistory:
    if i[0][0:2]=='12':
        dec=int(i[1])+dec
dec_daily=int(dec/31/10)
daily_averages.append("December,%s" %(dec_daily))


daily_averages=[i.split(',', 1) for i in daily_averages]
print(daily_averages)
```
