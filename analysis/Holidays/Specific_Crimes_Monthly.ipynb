# Comparing Offense Rates during Holiday Months

Retreive crime data and observe the number of times each offense is commited during months with a federal holiday:

```python
import sys
import pandas as pd
import csv


with open('crimes.csv') as csvfile: 
    crimes = list(csv.reader(csvfile, delimiter=',', quotechar='|'))

#clean and seperate columns:
for row in crimes:
    row=[i.strip() for i in row]
    row=[i.split(',', 1) for i in row]
    
#retrieve just report date and offense name:  
data=[]
for row in crimes:
    data.append([row[5], row[7]])
    
#get rid of header:
data.pop(0)
```

```python
#list of offenses:
offenses={'PETIT LARCENY':0,'HARRASSMENT 2':0,'ASSAULT 3 & RELATED OFFENSES':0,'CRIMINAL MISCHIEF & RELATED OF':0,'GRAND LARCENY':0,'DANGEROUS DRUGS':0,'OFF. AGNST PUB ORD SENSBLTY &':0,'ROBBERY':0,'BURGLARY':0,'FELONY ASSAULT':0}

#find the number of monthly crimes for each offense for the months that have federal holidays:
monthly_crimes = [{'PETIT LARCENY':0,'HARRASSMENT 2':0,'ASSAULT 3 & RELATED OFFENSES':0,'CRIMINAL MISCHIEF & RELATED OF':0,'GRAND LARCENY':0,'DANGEROUS DRUGS':0,'OFF. AGNST PUB ORD SENSBLTY &':0,'ROBBERY':0,'BURGLARY':0,'FELONY ASSAULT':0} for k in range(13)]

for row in data:
    if row[0][0:2] in {'01','02','05','07','09','10','11','12'}:
        if row[1] in offenses:
            monthly_crimes[int(row[0][0:2])][(row[1])]+=1


```

```python
#Get daily average for each month in question:
for offense in monthly_crimes[1]:
    monthly_crimes[1][offense]=int(monthly_crimes[1][offense]/31/10)
for offense in monthly_crimes[2]:
    monthly_crimes[2][offense]=int(monthly_crimes[2][offense]/28/10)
for offense in monthly_crimes[5]:
    monthly_crimes[5][offense]=int(monthly_crimes[5][offense]/31/10)
for offense in monthly_crimes[7]:
    monthly_crimes[7][offense]=int(monthly_crimes[7][offense]/31/10)
for offense in monthly_crimes[9]:
    monthly_crimes[9][offense]=int(monthly_crimes[9][offense]/30/10)
for offense in monthly_crimes[10]:
    monthly_crimes[10][offense]=int(monthly_crimes[10][offense]/31/10)
for offense in monthly_crimes[11]:
    monthly_crimes[11][offense]=int(monthly_crimes[11][offense]/30/10)
for offense in monthly_crimes[12]:
    monthly_crimes[12][offense]=int(monthly_crimes[12][offense]/31/10)

monthly_rates=[]
for month in [1,2,5,7,9,10,11,12]:
        monthly_rates.append(monthly_crimes[month])    
```

We observe how many crimes happened for each offense during holidays.

```python
#find the number of holiday crimes for each offense for the months that have federal holidays:
monthly_holidays = [{'PETIT LARCENY':0,'HARRASSMENT 2':0,'ASSAULT 3 & RELATED OFFENSES':0,'CRIMINAL MISCHIEF & RELATED OF':0,'GRAND LARCENY':0,'DANGEROUS DRUGS':0,'OFF. AGNST PUB ORD SENSBLTY &':0,'ROBBERY':0,'BURGLARY':0,'FELONY ASSAULT':0} for k in range(13)]

#retrieve list of federal holidays:
with open('holidays.csv') as csvfile: 
    holidays = list(csv.reader(csvfile, delimiter=',', quotechar='|'))

#record how many crimes per offense:
    for row in holidays:
            #looking at each row in data
            for row2 in data:
                #looking for a match:
                if row[1] == row2[0]:
                    if row2[1] in offenses:
                        monthly_holidays[int(row[1][0:2])][(row2[1])]+=1                     
```

```python
#Counts represent holiday over 10 year period. Must be divided daily.
for item in monthly_holidays:
    for offense in item:
        item[offense]=int(item[offense]/10)
        
#January and November each had 2 holidays. Thus we must take the average of their values:
for offense in monthly_holidays[1]:
    monthly_holidays[1][offense]=int((monthly_holidays[1][offense])/2)
        
for offense in monthly_holidays[11]:
    monthly_holidays[11][offense]=int((monthly_holidays[11][offense])/2)

holiday_rates=[]
for month in [1,2,5,7,9,10,11,12]:
        holiday_rates.append(monthly_holidays[month])

```
